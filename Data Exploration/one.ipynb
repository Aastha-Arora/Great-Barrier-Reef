```python
import pandas as pd
```

```python
# Davies Reef data from (15 station link)
# link documentation: https://coralreefwatch.noaa.gov/satellite/data_nrt/descriptions/timeseries.txt
# The day is collected 2 days of week (Mon and Thur) and the data is composite for preceding 3/4 days
data = pd.read_fwf('vs_ts_DaviesReef_GBR.txt')
data
```

```python
# Davis Reef Data for 2017
data_2017 = data.loc[(data['BYYY'] == 2017)]
data_2017.to_excel('data_2017.xlsx')
```

```python
# Central GBR Data (from 7 station link). This data has day wise SST
# link documentation: https://coralreefwatch.noaa.gov/product/vs/description.php#ascii

# Polygon Middle Longitude:
# 148.2750 
 
# Polygon Middle Latitude:
# -19.2250

# Averaged Maximum Monthly Mean:
# 28.3422

# Averaged Monthly Mean (Jan-Dec):
# 28.2080 28.3422 27.7395 26.6994 25.3221 23.6220 22.7473 22.7427 23.6184 25.0898 26.6232 27.483

data1 = pd.read_fwf('gbr_central.txt')
data1
```

```python
# Data for year 2017
data1_2017 = data1.loc[(data1['YYYY'] == 2017)]
data1_2017
```

```python
grouped = data1_2017.groupby('MM')
month_wise = grouped.mean()
month_wise
```
