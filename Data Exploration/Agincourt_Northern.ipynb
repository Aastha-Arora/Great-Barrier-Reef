```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Northern GBR Data (from 7 station link)

# Polygon Middle Longitude:
# 145.9750 
 
# Polygon Middle Latitude:
# -16.1000

data_northern = pd.read_fwf('NorthernGBR.txt')
data_northern.head()
```

```python
#Get mean of temperate points through grouped year
grouped_year = data_northern.groupby('YYYY')
mean_temp_grouped_year = grouped_year['SST_MAX'].mean()
mean_temp_grouped_year_df = pd.DataFrame(mean_temp_grouped_year)

#Resetting index
mean_temp_grouped_year_df = mean_temp_grouped_year_df.reset_index()
mean_temp_grouped_year_df

#Renaming column titles
mean_temp_grouped_year_df = mean_temp_grouped_year_df.rename(columns={'YYYY':'Years',
                              'SST_MAX':'Max SST'
                             })

mean_temp_grouped_year_df
```

```python
#Chart a scatterplot with Dataframe above
# x_axis: Year
# y_axis: Max SS temperature

#setting x-axis and y-axis
x_axis = mean_temp_grouped_year_df['Years']
y_axis = mean_temp_grouped_year_df['Max SST']

#create plot
plt.figure(figsize=(20,10))
plt.plot(x_axis, y_axis, marker="s", color="Red", linewidth=1, label="Temperature")

# Create labels for the x-axis and y-axis
plt.xlabel("Years")
plt.ylabel("Max Sea Surface Temperature")

plt.show()
```

```python
#ISSUE IS 4 MONTH DATA POINTS DURING SUMMER AND NOT FULL YEAR

# Agincourt Reff data from "Coral Bleaching Risk Indicators and Summer Sea Water Temperatures for the Great Barrier Reef"
#Link: http://data.aims.gov.au/aimsrtds/coralbleaching.xhtml?fromDate=2013%2C2014%2C2015%2C2016%2C2017%2C2018%2C2019&dataLevel=1

sst_agincourt_raw = pd.read_csv("AgincourtBleachingThresholds.csv")
sst_agincourt = sst_agincourt_raw.dropna(how='any')
```

```python
#Formatting table
#Rename titles

sst_agincourt_clean = sst_agincourt.rename(columns={'wtemp_LEVEL1_summer_13_14':'Nov 2013 - Mar 2014',
                              'wtemp_LEVEL1_summer_14_15':'Nov 2014 - Mar 2015',
                              'wtemp_LEVEL1_summer_15_16':'Nov 2015 - Mar 2016',
                              'wtemp_LEVEL1_summer_16_17':'Nov 2016 - Mar 2017',                             
                              'wtemp_LEVEL1_summer_17_18':'Nov 2017 - Mar 2018',
                              'wtemp_LEVEL1_summer_18_19':'Nov 2018 - Mar 2019',
                              'wtemp_LEVEL1_summer_19_20':'Nov 2019 - Mar 2020'
                             })
```

```python
#Get mean for temperatures in each year

#2013-2014
sst_agincourt_1314_ave = sst_agincourt_clean['Nov 2013 - Mar 2014']
sst_agincourt_1314_ave = sst_agincourt_1314_ave.mean()
#sst_agincourt_1314_ave


#2014-2015
sst_agincourt_1415_ave = sst_agincourt_clean['Nov 2014 - Mar 2015']
sst_agincourt_1415_ave = sst_agincourt_1415_ave.mean()
#sst_agincourt_1415_ave


#2015-2016
sst_agincourt_1516_ave = sst_agincourt_clean['Nov 2015 - Mar 2016']
sst_agincourt_1516_ave = sst_agincourt_1516_ave.mean()
#sst_agincourt_1516_ave

#2015-2016
sst_agincourt_1516_ave = sst_agincourt_clean['Nov 2015 - Mar 2016']
sst_agincourt_1516_ave = sst_agincourt_1516_ave.mean()
#sst_agincourt_1516_ave

#2016-2017
sst_agincourt_1617_ave = sst_agincourt_clean['Nov 2016 - Mar 2017']
sst_agincourt_1617_ave = sst_agincourt_1617_ave.mean()
#sst_agincourt_1617_ave


#2017-2018
sst_agincourt_1718_ave = sst_agincourt_clean['Nov 2017 - Mar 2018']
sst_agincourt_1718_ave = sst_agincourt_1718_ave.mean()
#sst_agincourt_1718_ave

#2018-2019
sst_agincourt_1819_ave = sst_agincourt_clean['Nov 2018 - Mar 2019']
sst_agincourt_1819_ave = sst_agincourt_1819_ave.mean()
#sst_agincourt_1819_ave

#2019-2020
sst_agincourt_1920_ave = sst_agincourt_clean['Nov 2019 - Mar 2020']
sst_agincourt_1920_ave = sst_agincourt_1920_ave.mean()
#sst_agincourt_1920_ave
```

```python
#Adding temperature means to Dataframe

sst_agincourt_mean_temp = {"Nov 2013 - Mar 2014":  [sst_agincourt_1314_ave],
                           "Nov 2014 - Mar 2015" : [sst_agincourt_1415_ave],
                           "Nov 2015 - Mar 2016" : [sst_agincourt_1516_ave],
                           "Nov 2016 - Mar 2017" : [sst_agincourt_1617_ave],
                           "Nov 2017 - Mar 2018" : [sst_agincourt_1718_ave],
                           "Nov 2018 - Mar 2019" : [sst_agincourt_1819_ave],
                           "Nov 2019 - Mar 2020" : [sst_agincourt_1920_ave]
                            }
sst_agincourt_mean_temp_df = pd.DataFrame(sst_agincourt_mean_temp, index = ["Mean Temperature"])
 
```

```python
#Changing index to column

sst_agincourt_mean_temp_df_transpose = sst_agincourt_mean_temp_df.T
sst_agincourt_mean_temp_df_transpose.index.name = "Year"
sst_agincourt_temp = sst_agincourt_mean_temp_df_transpose.reset_index()

sst_agincourt_temp
```

```python
#Chart a scatterplot with Dataframe above
# x_axis: date
# y_axis: temperature


#setting x-axis and y-axis
y_axis = sst_agincourt_temp['Mean Temperature']
x_axis = sst_agincourt_temp['Year']


#create plot
plt.figure(figsize=(20,10))
plt.plot(x_axis, y_axis, marker="s", color="Red", linewidth=1, label="Temperature")



# Create labels for the x-axis and y-axis
plt.xlabel("Years")
plt.ylabel("Average Sea Surface Temperature")

plt.show()
```
