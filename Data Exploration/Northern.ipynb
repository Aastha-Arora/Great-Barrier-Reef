```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Northern GBR Data (from 7 station link)
#Link: https://coralreefwatch.noaa.gov/product/vs/data.php

# Polygon Middle Longitude:
# 145.9750 
 
# Polygon Middle Latitude:
# -16.1000

#Read raw data text file
data_northern = pd.read_fwf(f'Data Exploration/NorthernGBR.txt')
```

```python
#Get mean of temperature points through grouped year for 2000-2020
grouped_year = data_northern.groupby('YYYY')
mean_temp_grouped_year = grouped_year['SST_MAX'].mean()
mean_temp_grouped_year_df = pd.DataFrame(mean_temp_grouped_year)

#Resetting index
mean_temp_grouped_year_df = mean_temp_grouped_year_df.reset_index()

#Renaming column titles
mean_temp_grouped_year_df = mean_temp_grouped_year_df.rename(columns={'YYYY':'Years',
                              'SST_MAX':'Max SST'
                             })
```

```python
#Chart a line plot for Average SST for 2000-2020

#setting x-axis and y-axis
x_axis_ave = mean_temp_grouped_year_df['Years']
y_axis_ave = mean_temp_grouped_year_df['Max SST']


#create plot
plt.figure(figsize=(20,10))

plt.xticks(x_axis_ave, fontsize="15")
plt.yticks(fontsize="15")

plt.plot(x_axis_ave, y_axis_ave, marker="o", color="Blue", linewidth=2, label="Max SS Temperature")

# Create labels for the x-axis, y-axis, and title
plt.title("Maximum Sea Surface Temperature and years from 2000-2020 for Northern GBR Station", fontsize="25")
plt.xlabel("Years", fontsize="20")
plt.ylabel("Max Sea Surface Temperature", fontsize="20")

plt.grid(True)

plt.savefig(f"Output/Average SST for 2000 2020.png")

plt.show()


```

```python
#Setting up for bar chart of Degree Heating Week values for years 2000-2020

#Link: https://coralreefwatch.noaa.gov/product/vs/description.php
#DHW_from_90th_HS>1:
#The daily Degree Heating Week value calculated by accumulating 
#daily 90th percentile HotSpot values (greater than 1) for pixels 
#contained in a Regional Virtual Station. This value is represented 
#as a solid red line on the time series graphs.

#Filtering for each year
grouped_year = data_northern.groupby('YYYY')

#Getting max amount for each year in the Degree Heating Week column
dhw_grouped_year = grouped_year['DHW_from_90th_HS>1'].max()
dhw_grouped_year_df = pd.DataFrame(dhw_grouped_year)

#Resetting index
dhw_grouped_year_df = dhw_grouped_year_df.reset_index()


#Renaming column titles
dhw_grouped_year_df = dhw_grouped_year_df.rename(columns = {'YYYY': 'Years',
                                                           'DHW_from_90th_HS>1' : 'Degree Heating Week Values'
                                                          })
dhw_grouped_year_df.head()
```

```python
#Creating a bar chart for Degree Heating Week values for 2000-2020

#Setting x-axis and y-axis
x_axis_dhw = dhw_grouped_year_df['Years']
y_axis_dhw = dhw_grouped_year_df['Degree Heating Week Values']

#Plotting chart
plt.figure(figsize=(20,10))

plt.xticks(x_axis_dhw , fontsize="15")
plt.yticks(fontsize = '15')

plt.bar(x_axis_dhw, y_axis_dhw, color='r', alpha=0.5, align="center")

#Labeling x-axis and y-axis and title
plt.title("Degree Heating Week Values for years between 2000-2020 for Northern GBR", fontsize="20")
plt.xlabel("Years", fontsize="20")
plt.ylabel("Degree Heating Week Values(C)", fontsize="20")

plt.grid(True)

plt.savefig(f"Output/Degree Heating Week values for 2000-2020.png")

plt.show()
```

```python
#Graphing Ave SST Max of 2000-2020 with histogram of Degree Heating Week values for years 
#2000-2020

#Creating a figure and axis objects with subplots
fig, ax = plt.subplots(figsize=(30,15))

#Making the first plot of bar chart Deagree Heating Week
ax.bar(dhw_grouped_year_df['Years'], dhw_grouped_year_df['Degree Heating Week Values'], label='Degree Heating Week Values',color='r', alpha=0.5, align="center")

#Setting x-axis and y-axis, tables, ticks, and title for bar chart
plt.xticks(dhw_grouped_year_df['Years'],fontsize="25")
plt.yticks(fontsize='20')

plt.xlabel("Years", fontsize="25")
ax.set_ylabel('Degree Heating Week Values', color='red', fontsize=30)

plt.title("Ave Max SST and Degree Heating Week Values for years between 2000-2020 for Northern GBR", fontsize="30")


#Using twin object for two different y-axis
ax2 = ax.twinx()

#Creating line plot for SST Max with different y-axis
ax2.plot(mean_temp_grouped_year_df['Years'], mean_temp_grouped_year_df['Max SST'],linewidth=5, label='Max Sea Surface Temperature')

#Setting label for line plot y-axis label and ticks
ax2.set_ylabel('Sea Surface Temperature(C)', color='blue', fontsize=30)
plt.yticks(fontsize='20')

#Setting the legend
plt.legend(loc="best", fontsize='20')

plt.savefig(f"Output/SSTMAX and DHW for 2000-2020.png")


plt.show()
```

```python
#Histogram for 2006, 2014 and 2010, 2016 to show the distribution 

#Function for creating histogram
def histogram_sst_max_function(year_value):
    
    histogram_sstmax = data_northern.loc[data_northern["YYYY"] == year_value]['SST_MAX']

    quartiles = histogram_sstmax.quantile([.25,.5,.75])
    lowerq= quartiles[0.25]
    upperq  = quartiles[0.75]
    iqr = upperq - lowerq

#Calculating and storing years for 2006, 2014 and 2010, 2016  with histogram function
histogram_sst_max_function(2006)
histogram_sstmax = histogram_sstmax_2006

histogram_sst_max_function(2010)
histogram_sstmax = histogram_sstmax_2010

histogram_sst_max_function(2014)
histogram_sstmax = histogram_sstmax_2014

histogram_sst_max_function(2016)
histogram_sstmax = histogram_sstmax_2016
```

```python
#Creating and formatting boxplot of data on SST Max for year 2006, 2014, 2010, 2016

#Creating figure size
plt.figure(figsize=(30,20))

#Formating boxplot
plt.xticks(fontsize='35')
plt.yticks(fontsize= '35')

plt.xlabel('Years', fontsize='35')
plt.ylabel('SST MAX Temperature(F)', fontsize='35')

plt.title('Max Sea Surface Temperature from 2006, 2010, 2014, 2016 for Northern GBR', fontsize='40')

boxprops = dict(linestyle='-', linewidth=4, color='k')
medianprops = dict(linestyle='-', linewidth=4, color='k')


#Plotting boxplot
#Boxplot color source: https://matplotlib.org/3.1.1/gallery/statistics/boxplot_color.html
#Source: https://stackoverflow.com/questions/28740234/face-pattern-for-boxes-in-boxplots
bplots = plt.boxplot([histogram_sstmax_2006, histogram_sstmax_2014,histogram_sstmax_2010, histogram_sstmax_2016], 
            labels=['2006', '2014', '2010', '2016'],
            boxprops=boxprops,
            medianprops=medianprops)


plt.grid(True)

plt.savefig(f"Output/SSTMAX boxplot for 2006-2014-2010-2016.png")

plt.show()
```
