```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Northern GBR Data (from 7 station link)
#Link: https://coralreefwatch.noaa.gov/product/vs/data.php

# Polygon Middle Longitude:
# 145.9750 
 
# Polygon Middle Latitude:
# -16.1000

data_northern = pd.read_fwf('NorthernGBR.txt') #[remember to fix]
#data_northern.head()
```

```python
#Get mean of temperature points through grouped year for 2000-2020
grouped_year = data_northern.groupby('YYYY')
mean_temp_grouped_year = grouped_year['SST_MAX'].mean()
mean_temp_grouped_year_df = pd.DataFrame(mean_temp_grouped_year)

#Resetting index
mean_temp_grouped_year_df = mean_temp_grouped_year_df.reset_index()
mean_temp_grouped_year_df

#Renaming column titles
mean_temp_grouped_year_df = mean_temp_grouped_year_df.rename(columns={'YYYY':'Years',
                              'SST_MAX':'Max SST'
                             })

#mean_temp_grouped_year_df
```

```python
#Chart a line plot for Average SST for 2000-2020
# x_axis: Year
# y_axis: Max SS temperature

#setting x-axis and y-axis
x_axis_ave = mean_temp_grouped_year_df['Years']
y_axis_ave = mean_temp_grouped_year_df['Max SST']

#create plot
plt.figure(figsize=(20,10))
plt.xticks(x_axis_ave, fontsize="15")
plt.yticks(fontsize="15")
plt.plot(x_axis_ave, y_axis_ave, marker="o", color="Blue", linewidth=2, label="Max SS Temperature")

# Create labels for the x-axis and y-axis
plt.xlabel("Years", fontsize="20")
plt.ylabel("Max Sea Surface Temperature", fontsize="20")
plt.title("Maximum Sea Surface Temperature and years from 2000-2020 for Northern GBR Station", fontsize="25")
plt.grid(True)

plt.show()


```

```python
#Get mean of temperature points through grouped year for 2015-2020
#Filtering for years 2015-2020
mean_grouped_year_1520 = data_northern.loc[(data_northern['YYYY'] > 2014) &
                                      (data_northern['YYYY'] <= 2020)]

#Grouping by years
mean_grouped_year_1520 = mean_grouped_year_1520.groupby('YYYY')


# Getting average for 2015-2020 in the MAX SST column
mean_grouped_year_1520 = mean_grouped_year_1520['SST_MAX'].mean()

#Creating a DataFrame
mean_grouped_year_1520_df = pd.DataFrame(mean_grouped_year_1520)

#Resetting index
mean_grouped_year_1520_df = mean_grouped_year_1520_df.reset_index()

#Renaming column titles
mean_grouped_year_1520_df = mean_grouped_year_1520_df.rename(columns={'YYYY':'Years',
                                                                      'SST_MAX':'Max SST'
                                                                       })
#mean_grouped_year_1520_df 
```

```python
#Chart a line plot for Average SST for 2015-2020
# x_axis: Year
# y_axis: Max SS temperature

#setting x-axis and y-axis
x_axis_ave_1520 = mean_grouped_year_1520_df['Years']
y_axis_ave_1520 = mean_grouped_year_1520_df['Max SST']

#create plot
plt.figure(figsize=(20,10))
plt.xticks(x_axis_ave_1520, fontsize="20")
plt.yticks(fontsize="20")
plt.plot(x_axis_ave_1520, y_axis_ave_1520, marker="o", color="Blue", linewidth=2, label="Max SS Temperature")

# Create labels for the x-axis and y-axis
plt.xlabel("Years", fontsize="20")
plt.ylabel("Max Sea Surface Temperature", fontsize="20")
plt.title("Maximum Sea Surface Temperature and years from 2015-2020 for Northern GBR Station", fontsize="25")
plt.grid(True)

plt.show()


```

```python
#SHOW AASTHA

#Graphing histogram of DHW_from_90th_HS>1 for years 2000-2020
#Link: https://coralreefwatch.noaa.gov/product/vs/description.php
#DHW_from_90th_HS>1:
#The daily Degree Heating Week value calculated by accumulating 
#daily 90th percentile HotSpot values (greater than 1) for pixels 
#contained in a Regional Virtual Station. This value is represented 
#as a solid red line on the time series graphs.

#Filtering for each year
grouped_year = data_northern.groupby('YYYY')

#Getting max amount for each year in the Degree Heating Week column
dhw_grouped_year = grouped_year['DHW_from_90th_HS>1'].max()
dhw_grouped_year_df = pd.DataFrame(dhw_grouped_year)

#Resetting index
dhw_grouped_year_df = dhw_grouped_year_df.reset_index()


#Renaming column titles
dhw_grouped_year_df = dhw_grouped_year_df.rename(columns = {'YYYY': 'Years',
                                                           'DHW_from_90th_HS>1' : 'Degree Heating Week Values'
                                                          })
#dhw_grouped_year_df

```

```python
#Graphing histogram of SST@90th_HS for years 2000-2020
#Link: https://coralreefwatch.noaa.gov/product/vs/description.php
#SST@90th_HS:
#The daily SST value where the HotSpot value is equal to the 90th 
#percentile HotSpot value for pixels contained in a Regional Virtual 
#Station. This acts as a moving pixel within a region that is free 
#to migrate with each data update. It follows thermal stress as 
#indicated by the HotSpot value for that day.

#Filtering for each year
grouped_year = data_northern.groupby('YYYY')

#Getting max amount for each year in the Degree Heating Week column
sst90_grouped_year = grouped_year['SST@90th_HS'].max()
sst90_grouped_year_df = pd.DataFrame(sst90_grouped_year)

#Resetting index
sst90_grouped_year_df = sst90_grouped_year_df.reset_index()


#Renaming column titles
sst90_grouped_year_df = sst90_grouped_year_df.rename(columns = {'YYYY': 'Years',
                                                           'SST@90th_HS' : '90th percentile HotSpot values'
                                                           })
```
